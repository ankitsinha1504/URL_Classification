```python
urls_df = pd.read_csv("urls.csv")
urls_df
```

```python
urls = urls_df["website_url"].tolist()
urls
```

```python
description = []
```

```python
#sample_demo
url = "https://www.ebookers.com/?AFFCID=EBOOKERS-UK.network.affiliatewindow.85386&awc=7904_1606029543_518c2af76500871f76cadec26f8eb055"
headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; Intel Mac OS X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
driver = webdriver.Chrome(PATH)
driver.get(url)
print(driver.title)
driver.close()
```

```python

for i in range(0 , len(urls)):
    url = urls[i]
    driver = webdriver.Chrome(PATH)
    try:
        driver.get(url)
        if(driver.title != None):
            print(f" {i} and {driver.title}")
            description.append(driver.title)
        else:
                description.append(None)
        driver.close()
    except (selenium.common.exceptions.WebDriverException , NameError , RuntimeError):
        print(f" {i} and Exception")
        description.append(None)
        driver.close()
```

```python
description
```

```python
df = pd.DataFrame(description , columns=["Titles"])
```

```python
df.to_csv("all_titles.csv", index = False)
```
